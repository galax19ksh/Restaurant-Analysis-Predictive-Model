### Task 2: Descriptive analysis

**2.1:** Statistical analysis of numerical columns

```python
#calculate statistical measures ie mean, median and standard deviation

numeric_columns = df[['Average Cost for two', 'Price range', 'Aggregate rating']]

def calculate_stats(df, columns):
    for col in columns:
        mean_val = df[col].mean()
        median_val = df[col].median()
        std_dev_val = df[col].std()

        print(f"STATISTICS FOR {col}:")
        print(f"Mean: {mean_val:.2f}")
        print(f"Median: {median_val:.2f}")
        print(f"Standard Deviation: {std_dev_val:.2f}")
        print()

calculate_stats(df, numeric_columns)
```
